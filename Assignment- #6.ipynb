```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.cluster.hierarchy as shc
from scipy.spatial.distance import squareform, pdist
```

```python
a = np.array([0.4005,0.2148,0.3457,0.2652,0.0789,0.4548])
b = np.array([0.5306,0.3854,0.3156,0.1875,0.4139,0.3022])

# the given 6 points
point = ['P1','P2','P3','P4','P5','P6']
data = pd.DataFrame({'Point':point, 'x cordinate':a, 'y cordinate':b})
#X-Y coordinate of 6 points
data = data.set_index('Point')
data
```

```python
#distance matrix for the given 6 points
dist = pd.DataFrame(squareform(np.round(pdist(data[['x cordinate', 'y cordinate']]),4), 'euclidean'), columns=data.index.values, index=data.index.values)
dist
```

```python
#dengogram with single linkage representation
plt.figure(figsize=(10,4)) 
plt.title("Dendrogram with SINGLE linkage")  
dend = shc.dendrogram(shc.linkage(data[['x cordinate', 'y cordinate']], method='single'), labels=data.index)
```

```python
#dengogram with complete linkage representation
plt.figure(figsize=(10,4)) 
plt.title("Dendrogram with COMPLETE linkage")  
dend = shc.dendrogram(shc.linkage(data[['x cordinate', 'y cordinate']], method='complete'), labels=data.index)
```

```python
#dengogram with average linkage representation
plt.figure(figsize=(10,4)) 
plt.title("Dendrogram with AVERAGE linkage")  
dend = shc.dendrogram(shc.linkage(data[['x cordinate', 'y cordinate']], method='average'), labels=data.index)
```

```python
#importing all required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing,metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering #It is a type of hierarchical clustering algorithm.
from sklearn.metrics import silhouette_score

import warnings
warnings.filterwarnings("ignore")
```

```python
# a) Preprocessing the data by removing the categorical column and filling the missing values.
#importing dataset using pandas
dataframe = pd.read_csv('CC GENERAL.csv')
dataframe.info()
```

```python
#statistical display of first five rows of a dataset
dataframe.head()
```

```python
#displays data description in a dataframe.
dataframe.describe()
```

```python
#first five rows of a dataframe after droppping CUST_ID column which is a categorical column
df = dataframe.drop(['CUST_ID'], axis=1)
df.head()
```

```python
#finding null values
df.isnull().any()
```

```python
#filling the missing values 
df.fillna(dataframe.mean(), inplace=True)
df.isnull().any()
```

```python
df.corr().style.background_gradient(cmap="Greens")
```

```python
# b) Applying StandardScaler() and normalize() functions to scale and normalize raw input data.
x = df.iloc[:,0:-1]
y = df.iloc[:,-1]

#applying standard scalar to scale the data
scaler = preprocessing.StandardScaler()
scaler.fit(x)
X_scaled_array = scaler.transform(x)
X_scaled_df = pd.DataFrame(X_scaled_array, columns = x.columns)
```

```python
#Basically normalization is the process of scaling individual samples to have unit norm. 
#This process will be useful if planned to use quadratic form like dot-product or 
 #any other kernel to quantify the similarity of any pair of samples.
X_normalized = preprocessing.normalize(X_scaled_df) #data normalization of scaled data
# numpy array conversion into pandas dataframe
X_normalized = pd.DataFrame(X_normalized)
```

```python
# c) Using PCA with K=2 to reduce the input dimensions to two features.
#Applying PCA with k=2 to reduce the data dimensionality to two.
pca2 = PCA(n_components=2)
principalComponents = pca2.fit_transform(X_normalized)

principalDf = pd.DataFrame(data = principalComponents, columns = ['P1', 'P2'])

finalDf = pd.concat([principalDf, df[['TENURE']]], axis = 1)
finalDf.head()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(finalDf['P1'],finalDf['P2'],c=finalDf['TENURE'],cmap='prism', s =5)
plt.xlabel('pc1')
plt.ylabel('pc2')
```

```python
# d)Apply Agglomerative Clustering with k=2,3,4 & 5 on reduced features and visualize result for each k value using scatter plot
#applying agglomerative clustering with k=2 
ac2 = AgglomerativeClustering(n_clusters = 2)
 
# Clustering visualization
plt.figure(figsize =(6, 6))
plt.scatter(principalDf['P1'], principalDf['P2'], c = ac2.fit_predict(principalDf), cmap ='rainbow')
plt.show()
```

```python
#applying agglomerative clustering with k=3
ac3 = AgglomerativeClustering(n_clusters = 3)
 
# Clustering visualization
plt.figure(figsize =(6, 6))
plt.scatter(principalDf['P1'], principalDf['P2'], c = ac3.fit_predict(principalDf), cmap ='rainbow')
plt.show()
```

```python
#applying agglomerative clustering with k=4
ac4 = AgglomerativeClustering(n_clusters = 4)
 
# Clustering visualization
plt.figure(figsize =(6, 6))
plt.scatter(principalDf['P1'], principalDf['P2'], c = ac4.fit_predict(principalDf), cmap ='rainbow')
plt.show()
```

```python
#applying agglomerative clustering with k=5
ac5 = AgglomerativeClustering(n_clusters = 5)
 
# Clustering visualization
plt.figure(figsize =(6, 6))
plt.scatter(principalDf['P1'], principalDf['P2'], c = ac5.fit_predict(principalDf), cmap ='rainbow')
plt.show()
```

```python
# e) Evaluate different variations using Silhouette Scores and Visualize results with a bar chart.
k = [2, 3, 4, 5]
 
# for different values of k,appending the Silhouette scores of the different models(k values) to the list
silhouette_scores = []
silhouette_scores.append(silhouette_score(principalDf, ac2.fit_predict(principalDf)))
silhouette_scores.append(silhouette_score(principalDf, ac3.fit_predict(principalDf)))
silhouette_scores.append(silhouette_score(principalDf, ac4.fit_predict(principalDf)))
silhouette_scores.append(silhouette_score(principalDf, ac5.fit_predict(principalDf)))
 
#Comparing the results by plotting bar graph
plt.bar(k, silhouette_scores)
plt.xlabel('no. of clusters', fontsize = 20)
plt.ylabel('S(i)', fontsize = 20)
plt.show()
```
